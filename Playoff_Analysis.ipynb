### Useful import

```python
import pandas as pd
import numpy as np
import glob
from collections import Counter
```

```python
path_playoff_summary = "Data/Playoff/summary/*"
path_playoff_goals = "Data/Playoff/goals/*"
path_playoff_shots = "Data/Playoff/shots/*"
path_playoff_shootouts = "Data/Playoff/shootouts/*"
path_playoff_penalties = "Data/Playoff/penalties/*"

dictPos = {'Stürmer' : 'Forward',
       'Verteidiger' : 'Defender'}

l = []
for file in glob.iglob(path_playoff_summary):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[21:26]
    df['Number of year'] = 1
    l.append(df)

playoff_summary = pd.concat(l)
playoff_summary.replace({"Pos" : dictPos}, inplace=True)

l = []
for file in glob.iglob(path_playoff_goals):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[19:24]
    l.append(df)

playoff_goals = pd.concat(l)
playoff_goals.replace({"Pos" : dictPos}, inplace=True)

l = []
for file in glob.iglob(path_playoff_shots):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[19:24]
    l.append(df)

playoff_shots = pd.concat(l)
playoff_shots.replace({"Pos" : dictPos}, inplace=True)

l = []
for file in glob.iglob(path_playoff_shootouts):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[23:28]
    l.append(df)

playoff_shootouts = pd.concat(l)
playoff_shootouts.replace({"Pos" : dictPos}, inplace=True)

l = []
for file in glob.iglob(path_playoff_penalties):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[23:28]
    l.append(df)

playoff_penalties = pd.concat(l)
playoff_penalties.replace({"Pos" : dictPos}, inplace=True)

playoff_summary.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_summary.sort_index(inplace=True)
playoff_summary.drop(playoff_summary.columns[[0]], axis=1, inplace=True)

playoff_shots.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_shots.sort_index(inplace=True)
playoff_shots.drop(playoff_shots.columns[[0]], axis=1, inplace=True)

playoff_shootouts.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_shootouts.sort_index(inplace=True)
playoff_shootouts.drop(playoff_shootouts.columns[[0]], axis=1, inplace=True)

playoff_goals.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_goals.sort_index(inplace=True)
playoff_goals.drop(playoff_goals.columns[[0]], axis=1, inplace=True)

playoff_penalties.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_penalties.sort_index(inplace=True)
playoff_penalties.drop(playoff_penalties.columns[[0]], axis=1, inplace=True)

player_info = pd.concat([playoff_summary, playoff_goals, playoff_shots, playoff_penalties], axis=1).sort_index()


player_info.replace({"Pos" : dictPos}, inplace=True)

player_info = player_info.loc[:,~player_info.columns.duplicated()]
player_info = player_info.reset_index().set_index(['Spieler','Pos','Team','year'])

player_info.to_csv('Data/Cleaned_Data/playoff_clean.csv')

player_info
```
