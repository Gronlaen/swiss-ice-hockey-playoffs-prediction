```python
import pandas as pd
import numpy as np
import glob
from collections import Counter
```

```python
path_playoff_summary = "Data/Playoff/summary/*"
path_playoff_goals = "Data/Playoff/goals/*"
path_playoff_shots = "Data/Playoff/shots/*"
path_playoff_shootouts = "Data/Playoff/shootouts/*"
path_playoff_penalties = "Data/Playoff/penalties/*"


l = []
for file in glob.iglob(path_playoff_summary):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[21:26]
    df['Number of year'] = 1
    l.append(df)

playoff_summary = pd.concat(l)

l = []
for file in glob.iglob(path_playoff_goals):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[19:24]
    l.append(df)

playoff_goals = pd.concat(l)

l = []
for file in glob.iglob(path_playoff_shots):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[19:24]
    l.append(df)

playoff_shots = pd.concat(l)

l = []
for file in glob.iglob(path_playoff_shootouts):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[23:28]
    l.append(df)

playoff_shootouts = pd.concat(l)

l = []
for file in glob.iglob(path_playoff_penalties):
    df = pd.read_csv(file, sep=';')
    df['year'] = file[23:28]
    l.append(df)

playoff_penalties = pd.concat(l)

playoff_summary.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_summary.sort_index(inplace=True)
playoff_summary.drop(playoff_summary.columns[[0]], axis=1, inplace=True)

playoff_shots.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_shots.sort_index(inplace=True)
playoff_shots.drop(playoff_shots.columns[[0]], axis=1, inplace=True)

playoff_shootouts.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_shootouts.sort_index(inplace=True)
playoff_shootouts.drop(playoff_shootouts.columns[[0]], axis=1, inplace=True)

playoff_goals.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_goals.sort_index(inplace=True)
playoff_goals.drop(playoff_goals.columns[[0]], axis=1, inplace=True)

playoff_penalties.set_index(['Spieler','Team','Pos','year','GP'], inplace=True)
playoff_penalties.sort_index(inplace=True)
playoff_penalties.drop(playoff_penalties.columns[[0]], axis=1, inplace=True)

player_info = pd.concat([playoff_summary, playoff_goals, playoff_shots, playoff_penalties], axis=1).sort_index()

dictPos = {'Stürmer' : 'Forward',
       'Verteidiger' : 'Defender'}
player_info.replace({"Pos" : dictPos}, inplace=True)

player_info = player_info.loc[:,~player_info.columns.duplicated()]
player_info = player_info.reset_index().set_index(['Spieler','Pos','Team','year'])

player_info.to_csv('Data/Cleaned_Data/playoff_clean.csv')

player_info
```

```python
playoff_summary_grouped = player_info.groupby(level=[0,2]).agg({"GP": "sum",
                                      "G": "sum",
                                      "A": "sum",
                                      "SOG": "sum",
                                      "PTS": "sum",
                                      "PIM Total": "sum",
                                      "Number of year": "sum"})
playoff_summary_grouped["PO_PTS"] = playoff_summary_grouped["PTS"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_G"] = playoff_summary_grouped["G"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_A"] = playoff_summary_grouped["A"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_SOG"] = playoff_summary_grouped["SOG"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_PIM"] = playoff_summary_grouped["PIM Total"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_PTS"] = playoff_summary_grouped["PTS"]/playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_GP"] = playoff_summary_grouped["GP"]
playoff_summary_grouped["PO_NoY"] = playoff_summary_grouped["Number of year"]
playoff_summary_grouped.drop(['PTS','G','A','SOG','PIM Total','GP','Number of year'], axis=1, inplace=True)
```

```python
df_regular = pd.DataFrame.from_csv('regular_season_clean.csv')
df_regular = df_regular.reset_index().set_index(['Spieler','Pos','Team'])
df_regular

df_regular_grouped = df_regular.groupby(level=[0,2]).agg({"GP": "sum",
                                      "G": "sum",
                                      "A": "sum",
                                      "SOG": "sum",
                                      "PTS": "sum",
                                      "PIM Total": "sum",
                                      "Number of year": "sum"})
df_regular_grouped["RE_PTS"] = df_regular_grouped["PTS"]/df_regular_grouped["GP"]
df_regular_grouped["RE_G"] = df_regular_grouped["G"]/df_regular_grouped["GP"]
df_regular_grouped["RE_A"] = df_regular_grouped["A"]/df_regular_grouped["GP"]
df_regular_grouped["RE_SOG"] = df_regular_grouped["SOG"]/df_regular_grouped["GP"]
df_regular_grouped["RE_PIM"] = df_regular_grouped["PIM Total"]/df_regular_grouped["GP"]
df_regular_grouped["RE_PTS"] = df_regular_grouped["PTS"]/df_regular_grouped["GP"]
df_regular_grouped["RE_GP"] = df_regular_grouped["GP"]
df_regular_grouped["RE_NoY"] = df_regular_grouped["Number of year"]
df_regular_grouped.drop(['PTS','G','A','SOG','PIM Total','GP','Number of year'], axis=1, inplace=True)
```

```python
common_grouped = df_regular_grouped[(df_regular_grouped.index.isin(playoff_summary_grouped.index))]

common_concat = pd.concat([common_grouped,playoff_summary_grouped], axis=1)

common_concat.dropna(inplace=True)

common_concat
```
