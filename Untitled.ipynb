```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
sns.set()
%matplotlib inline
```

```python
PO_path = 'Data/Team/Rankings/by_match/playoff/*'
df_match = []

for file in glob.iglob(PO_path):
    f = pd.read_csv(file, ';')
    f['Year'] = file[36:40]
    df_match.append(f)

df_match = pd.concat(df_match)
df_match = df_match[['Home', 'Away', 'Resultat', 'Year']]
df_match['Result'] = [ int(x[1] > x[3]) for x in df_match["Resultat"]]
df_match.drop('Resultat', axis=1, inplace=True)
df_match
```
